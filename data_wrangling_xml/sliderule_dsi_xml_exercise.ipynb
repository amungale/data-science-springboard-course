## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
# print names of all countries
for child in document_tree.getroot():
    print(child.find('name').text)
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    print('* ' + element.find('name').text + ':')
    capitals_string = ''
    for subelement in element.getiterator('city'):
        capitals_string += subelement.find('name').text + ', '
    print(capitals_string[:-2])
```

## XML exercise

Using data in 'data/mondial_database.xml', the examples above, and refering to https://docs.python.org/2.7/library/xml.etree.elementtree.html, find

1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

```python
from xml.etree import ElementTree as ET

document = ET.parse( './data/mondial_database.xml' )
root = document.getroot()
```

```python
root.tag
root.attrib
for child in root:
    print(child.tag, child.attrib)
```

```python
for element in root.findall('country'):
    print(element.attrib)
```

# 10 countries with the lowest infant mortality rates

```python
#Find all element tags
for element in root.iter():
    print(element.tag)
```

```python
for element in document.iterfind('country'):
    print('* ' + element.find('infant_mortality').text + ':')
```

```python
#Create dictionary store mortality rate by country
mr_dict = {}
for element in document.iterfind('country'):
    try:
        name = element.find('name').text 
        mr = element.find('infant_mortality').text
        mr_dict[name] = mr
    except:
        pass

#convert to pandas df    
import pandas as pd
mr_df = pd.DataFrame.from_dict(mr_dict, orient='index', dtype=None)
mr_df.columns = ['mortality rate']
mr_df.sort_values('mortality rate')
#Currently sorting numbers alphabetically, convert numbers to text
mr_df['mortality rate'] = pd.to_numeric(mr_df['mortality rate'])
mr_df.sort_values('mortality rate').head(10)
```

# 10 cities with the largest population

```python
#Create dictionary store population rate by city
pop_dict = {}
# print names of all countries and their mortality rates
for element in document.iterfind('country'):
    for subelement in element.getiterator('city'):
        try:
            name = subelement.find('name').text
            pop = subelement.find('population').text
            pop_dict[name] = int(pop)
        except:
            pass
        

pop_df = pd.DataFrame.from_dict(pop_dict, orient='index', dtype=None)
pop_df.columns = ['city population']
pop_df.sort_values('city population').tail(10)
```

# 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

```python
#Create dictionary store ethnic groups and corresponding populations
eg_dict = {}
# print names of all countries and their mortality rates
for element in document.iterfind('country'):
    try:
        name = element.find('ethnicgroup').text
        pop = int(element.find('population').text)
        eg_dict[name] = pop
    except:
        pass
        
eg_df = pd.DataFrame.from_dict(eg_dict, orient='index')
eg_df = eg_df.reset_index()
eg_df.columns = ['ethnicity','population']
eg_df = eg_df.groupby('ethnicity')['population'].sum()
eg_df.nlargest(10)
#eg_df.sort_values('population', ascending=False).head(10)
```

# name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
#Store river data in a dictionary
river_dict = {}

for river in root.findall('river'):
    try:
        river_name = river.find('name').text
        river_country = river.attrib['country']
        river_length = int(river.find('length').text)
        river_dict[river_name] = [river_country,river_length]
    except:
        pass

river_df = pd.DataFrame.from_dict(river_dict, orient='index')
river_df = river_df.reset_index()
river_df.columns = ['River Name', 'Countries', 'Length']
river_df.sort_values('Length').tail(1)
```
